# Splitting a Col into Multiple Dummies

We can convert our reasons for absence variables into Dummy variables. 

**Dummy Variable**: an explanatory binary variable that equals 1 if a certain categorical effect is present, and equals 0 if that same effect is absence.

### .get_dummies()

In Pandas, this method converts categorical variables into dummy variables.

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_columns
```

Check and see if reason_col is missing any variables. Store the results in a new column called check. 

**Legend:**

0: missing value
1: single value
2,3,4...: not possible, employees are absent for a single reason only

```python
reason_columns['check'] = reason_cols.sum(axis = 1) - 2
```

```python
reason_cols = reason_columns
reason_cols
```

```python
reason_cols['check'].unique() # check the number of unique values in our dummy variable column
```

```python
reason_cols['check'].sum(axis=0) # check to see the amount of values in the check column
```

```python
reason_cols = reason_columns.drop(['check'], axis=1)
```

```python
reason_cols
```
